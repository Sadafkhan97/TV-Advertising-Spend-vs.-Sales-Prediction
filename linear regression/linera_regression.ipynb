```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/Advertising.csv")
```

```python
df.head()
```

```python
plt.scatter(df['TV'],df['sales'])
plt.xlabel('TV')
plt.ylabel('sales')
plt.show()
```

```python
x = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# create an object of linear regression function as lr
lr = LinearRegression()
```

```python
# obove created object have some methods and attributes such as .fit
lr.fit(X_train,y_train)
```

```python
X_test
```

```python
y_test
```

```python
lr.predict(X_test.iloc[0:1])
```

```python
plt.scatter(df['TV'],df['sales'])
plt.plot(X_train,lr.predict(X_train),color="red")
plt.xlabel('TV')
plt.ylabel('sales')
plt.show()
```

```python
plt.scatter(df['TV'],df['sales'])
plt.plot(X_test,lr.predict(X_test),color="red")
plt.xlabel('TV')
plt.ylabel('sales')
plt.show()
```

```python
plt.scatter(df['TV'],df['sales'])
plt.plot(df['TV'],lr.predict(x),color="red")
plt.xlabel('TV')
plt.ylabel('sales')
plt.show()
```

```python
# y = mx + b
m = lr.coef_
print(m)
```

```python
 b = lr.intercept_
 print(b)
```

```python
y = m*X_test + b
print(y)
```
